# Deep Learning Text Summarization

This notebook demonstrates the text summarization and feedback generation.


```python
import sys
sys.path.append('..')

from src.dl_summarizer import TextSummarizer

# Initialize summarizer
summarizer = TextSummarizer()

# Sample text for summarization
sample_text = """
Artificial Intelligence is a branch of computer science that aims to create intelligent machines. 
These machines can perform tasks that typically require human intelligence, such as visual perception, 
speech recognition, decision-making, and language translation. Machine Learning is a subset of AI that 
enables systems to learn and improve from experience without being explicitly programmed. Deep Learning, 
a further subset of Machine Learning, uses neural networks with multiple layers to analyze various 
factors of data. Natural Language Processing allows computers to understand, interpret, and generate 
human language in a valuable way.
"""

print("Original Text:")
print(sample_text)
print(f"\nLength: {len(sample_text)} characters")
```

```python
# Summarize text
summary = summarizer.summarize_text(sample_text, target_length=100)

print("\nSummarized Text:")
print(summary)
print(f"\nLength: {len(summary)} characters")
print(f"Compression Ratio: {len(summary)/len(sample_text):.2%}")
```

```python
# Test feedback generation
print("\nFeedback Generation Examples:")
print("\nGood Performance:")
print(summarizer.generate_feedback('Mathematics', 'good'))

print("\nMedium Performance:")
print(summarizer.generate_feedback('Science', 'medium'))

print("\nNeeds Improvement:")
print(summarizer.generate_feedback('History', 'needs_improvement'))
```
